```python
# Exploratory Data Analysis on Merged Dataset (Univariate, Bivariate and multivariate analysis)

# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

```

```python
# import merged dataset
merged_data = pd.read_csv("C:/Users/HP/Desktop/Dafe/DS_Projects/FNOL_Insurance/data/processed/merged_data.csv")
```

```python
# Commence EDA with the merged dataset

# Analysis of the ultimate claim amount
plt.figure(figsize=(10,6))
sns.histplot(merged_data['Ultimate_Claim_Amount'], bins=50, kde=True)
plt.title('Distribution of Ultimate Claim Amount')
plt.xlabel('Ultimate Claim Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Univariate analysis of categorical variables
categorical_cols = ['Severity_Band', 'Occupation', 'ThirdParty_Role', 'TP_Injury_Severity', 'Claim_Type', 'Claim_Complexity', 'Region', 'Vehicle_Type']
for col in categorical_cols:
    plt.figure(figsize=(8,5))
    sns.countplot(data=merged_data, x=col, order=merged_data[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```
