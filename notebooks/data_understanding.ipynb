```python
# Import necessary Libraries and Dataset

import pandas as pd
import numpy as np

claims = pd.read_csv("C:/Users/HP/Desktop/Dafe/DS_Projects/FNOL_Insurance/data/raw/claims.csv")
policyholders = pd.read_csv("C:/Users/HP/Desktop/Dafe/DS_Projects/FNOL_Insurance/data/raw/policyholders.csv")
third_party = pd.read_csv("C:/Users/HP/Desktop/Dafe/DS_Projects/FNOL_Insurance/data/raw/third_parties.csv")

# Display the first few rows of each dataset to understand their structure
print("Claims Dataset:")
print(claims.head())
print("\nPolicyholders Dataset:")
print(policyholders.head())
print("\nThird Party Dataset:")
print(third_party.head())
```

```python
# Checks for missing values

claims_missing = claims.isnull().sum()
policyholders_missing = policyholders.isnull().sum()
third_party_missing = third_party.isnull().sum()
```

```python
print("\nMissing Values in Claims Dataset:")
print(claims_missing)

print("\nMissing Values in Policyholders Dataset:")
print(policyholders_missing)

print("\nMissing Values in Third Party Dataset:")
print(third_party_missing)
```

```python
# Check data types of each column

print("\nData Types in Claims Dataset:")
print(claims.dtypes)

print("\nData Types in Policyholders Dataset:")
print(policyholders.dtypes)

print("\nData Types in Third Party Dataset:")
print(third_party.dtypes)
```

```python
# List columns in each dataset

print("\nColumns in Claims Dataset:")
print(claims.columns)
```

```python
# List of columns in Policyholders Dataset

print("\nColumns in Policyholders Dataset:")
print(policyholders.columns)
```

```python
# List of columns in Policyholders Dataset

print("\nColumns in Policyholders Dataset:")
print(policyholders.columns)
```

```python
# Convert date columns to datetime format in the claims table

claims['Accident_Date'] = pd.to_datetime(claims['Accident_Date'])
claims['FNOL_Date'] = pd.to_datetime(claims['FNOL_Date'])
claims['Settlement_Date'] = pd.to_datetime(claims['Settlement_Date'])
```

```python
# Clean Claims table and fill in missing values

claims['Ultimate_Claim_Amount'].fillna(claims['Ultimate_Claim_Amount'].median(), inplace=True)
claims['Settlement_Date'].fillna(method='ffill', inplace=True)

claims.isnull().sum()
```

```python
# Merge the datasets on the common key 'Policy_ID'

merged_data = pd.merge(claims, policyholders, on='Policy_ID', how='left')
merged_data = pd.merge(merged_data, third_party, on='Claim_ID', how='left')

# Display the first few rows of the merged dataset
print("\nMerged Dataset:")
print(merged_data.head())
```

```python
merged_data.head()
```

```python
# Check for missing values in the merged dataset

merged_data.isnull().sum()
```

```python
merged_data.shape
```

```python
# drop duplicates if any

merged_data.drop_duplicates(inplace=True)
```

```python
merged_data['ThirdParty_Role'].value_counts()
```

```python
merged_data['TP_Injury_Severity'].value_counts()
```

```python
# Drop TP_ID

merged_data.drop(columns=['TP_ID'], inplace=True)
```

```python
# Fill NaN values in 'ThirdParty_Role' with 'Unknown' and in 'TP_Injury_Severity' with 'No Injury'

merged_data['ThirdParty_Role'].fillna('Unknown', inplace=True)
merged_data['TP_Injury_Severity'].fillna('No Injury', inplace=True)


# Check missing values again

merged_data.isnull().sum()
```

```python
# Clean merged_data if necessary

merged_data = merged_data[merged_data['Ultimate_Claim_Amount'] >= 0]
print("\nFinal Merged Dataset:")
print(merged_data.head())
```
